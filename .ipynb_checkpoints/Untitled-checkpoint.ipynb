```python
euc_lisa.head()
```

```python
edges = ox.graph_to_gdfs(G_projected, nodes=False, edges=True)
```

```python
quadrant1 = euc_lisa[euc_lisa["quadrant"] == 1]
```

```python
buffered_area = quadrant1.buffer(1000)  # 1000 metros = 1 km
buffered_union = buffered_area.union_all()
```

```python
edges_near_polygons = edges[edges.intersects(buffered_union)]
```

```python
ax = edges.plot(color="gray", linewidth=0.5)  # Todas las carreteras en gris
edges_near_polygons.plot(ax=ax, color="red", linewidth=0.5)  # Las que cumplen en rojo
quadrant1.plot(ax=ax, color="blue", alpha=0.3)
```

```python
fig, ax = plt.subplots(figsize=(12, 12))  # Cambia los valores para ajustar el tamaño

# Graficar carreteras
edges.plot(ax=ax, color="gray", linewidth=0.5)  # Todas las carreteras en gris
edges_near_polygons.plot(ax=ax, color="red", linewidth=0.5)  # Carreteras cercanas en rojo
quadrant1.plot(ax=ax, color="blue", alpha=0.3)  # Polígonos en azul

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
edges.plot(ax=ax, color="gray", linewidth=0.5)
edges_near_polygons.plot(ax=ax, color="red", linewidth=0.5)
quadrant1.plot(ax=ax, color="blue", alpha=0.3)

ax.set_xlabel("Coordenada X (metros o grados)")
ax.set_ylabel("Coordenada Y (metros o grados)")

plt.show()
```

```python
print(edges_near_polygons[["highway", "name", "length"]])
```

```python
# 1️⃣ Eliminar las carreteras seleccionadas del GeoDataFrame
edges_remaining = edges[~edges.index.isin(edges_near_polygons.index)]  # Filtrar las carreteras no seleccionadas
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
edges_remaining.plot(ax=ax, color="gray", linewidth=0.5)

quadrant1.plot(ax=ax, color="blue", alpha=0.3)

ax.set_xlabel("Coordenada X (metros o grados)")
ax.set_ylabel("Coordenada Y (metros o grados)")

plt.show()
```
