```python
ox.plot_graph(G_projected, node_color="r")
```

```python
# No necesitamos los nodos, solo las lineas:
edges = ox.graph_to_gdfs(G_projected, nodes=False, edges=True)
```

```python
# Vemos la estructura de los datos para ver el nombre del resultado del LISA. En este caso es el atributo: 'quadrant'
euc_lisa.head()
```

```python
quadrant1 = euc_lisa[euc_lisa["quadrant"] == 1]
```

```python
buffered_area = quadrant1.buffer(1000)  # 1000 metros = 1 km
buffered_union = buffered_area.union_all()
```

```python
edges_near_polygons = edges[edges.intersects(buffered_union)]
```

```python
fig, ax = plt.subplots(figsize=(12, 12)) # hacemos el lienzo grande
edges.plot(ax=ax, color="gray", linewidth=0.5) #representación de las carreteras en general
edges_near_polygons.plot(ax=ax, color="red", linewidth=0.5) # representación de las carreteras a menos de 1km
quadrant1.plot(ax=ax, color="blue", alpha=0.3) #polígonos con alta concentracion de eucaliptos.

ax.set_xlabel("Coordenada X (metros o grados)")
ax.set_ylabel("Coordenada Y (metros o grados)") 

plt.show()
```

Podemos hacer una lista de las carreteras cortadas, ya sea para informar a la población, o para posteriores procesos.

```python
print(edges_near_polygons[["highway", "name", "length"]])
```

Ahora vamos a crear un objeto sin las carreteras de riesgo, para poder generar rutas alternativas.

```python
edges_remaining = edges[~edges.index.isin(edges_near_polygons.index)]  
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
edges_remaining.plot(ax=ax, color="gray", linewidth=0.5)

quadrant1.plot(ax=ax, color="blue", alpha=0.3)

ax.set_xlabel("Coordenada X (metros o grados)")
ax.set_ylabel("Coordenada Y (metros o grados)")

plt.show()
```

Para crear rutas alternativas, utilizariamos el objeto edges_remaining. El índice de correlación de LISA no es el factor mas indicado para medir la posibilidad de incendios de manera realista, pero para un caso hipotetico nos sirve, ya que identificamos los polígonos con mas eucaliptos juntos. Se podrían integrar datos auxiliares como un inventario de biomasa forestal, o algo por el estilo, y conseguir identificar las masas boscosas con mas riesgo de incendio para seleccionar esas carreteras, pero en la práctica sería repetir lo que hemos hecho.
